```python
from GoogleNews import GoogleNews
googlenews = GoogleNews()
googlenews.set_lang('en')
```

```python
# Get last 5 years Amazon's google news
googlenews.get_news('Amazon')  
googlenews.set_time_range('01/01/2016','12/31/2020')  
```

```python
results=googlenews.results()
results
```

```python
headlines=googlenews.get_texts()
headlines

   
```

```python
import pandas as pd
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer() #Creating a SentimentIntensityAnalyzer object
columns = [ 'Headline']
NewsHeadlines = pd.DataFrame(headlines, columns=columns)
NewsHeadlines
```

```python
# Creating sentiment scores dictionary using polarity_scores method of SentimentIntensityAnalyzer
# Resultant dictionary contains pos, neg, neu, and compound scores.
sentiment_scores = NewsHeadlines['Headline'].apply(analyzer.polarity_scores).tolist()
sentiment_scores = pd.DataFrame(sentiment_scores)
print("Overall sentiment dictionary is :\n  ")
print(sentiment_scores)
```

```python
 
sentiment_value=[] 
 # decide sentiment as positive, negative and neutral
for i in  sentiment_scores['compound']:
   
    if  i >= 0.05  :
        sentiment_value.append( "Positive")
 
    elif i  <= - 0.05 :
        sentiment_value.append( "Negative")
 
    else :
        sentiment_value.append("Neutral") 
   
   
```

```python
df = pd.DataFrame( NewsHeadlines['Headline'] )
```

```python
df['Sentiment value']=(sentiment_value)
df
```
